##FILTRO 1

**DETENCION DE BORDES**

Los bordes de una imagen digital se pueden definir como transiciones entre dos regiones de
niveles de gris significativamente distintos. Suministran una valiosa información sobre las
fronteras de los objetos y puede ser utilizada para segmentar la imagen, reconocer objetos, etc.
La mayoría de las técnicas para detectar bordes emplean operadores locales basados en distintas
aproximaciones discretas de la primera y segunda derivada de los niveles de grises de la imagen.

**En este filtro el objetivo principal es detectar bordes desplegando un color blanco para ver los limites de estos.**

````
import cv2
from google.colab.patches import cv2_imshow

imagen=cv2.imread('/content/guitarra.jpg')
imagen=cv2.cvtColor(imagen,cv2.COLOR_RGB2BGR)
filtro_canny=cv2.Canny(imagen,60,150)

contours,contornos=cv2.findContours(filtro_canny,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(imagen,contours,-2,(255,0,0),5)


cv2_imshow(filtro_canny)

cv2_imshow(imagen)
````

````
def DetectarForma_a_(contorno, width, height):
  epsilon = 0.01*cv2.arcLength(contorno, True)
  approx = cv2.aprroxPolyDP(contorno, epsilon, True)

  if len(approx) == 3:
    forma ='Triangulo'
  elif len(approx) == 4:
    escala = float(width)/height
    if escala == 1:
      forma = 'Cuadrado'
    else:
      forma = 'Rectangulo'
  elif len(approx) == 5:
    forma = 'Pentagono'
  elif len(approx) == 6:
    forma = 'Hexagono'
  elif len (approx) == 10:
    forma = 'Circulo'

    return forma
````

##Tracking usando OpenCV

**El tracking** Es una funcion que nos permite, detectar el movimiento de objetos, personas,etc. Y podemos seguirlos o rastrearlos.

````
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
````

````
Captura = cv2.VideoCapture('/content/e-coli-swimming1.avi')

fourcc = cv2.VideoWriter_fourcc('X','V','I','D')
VideoSalida = cv2.VideoWriter("MovimientoDetectadoE-Coli.mp4", fourcc, 5.0, (1280, 720))

while Captura.read()[0] == True:
    Estado, Frame1 = Captura.read()
    Estado, Frame2 = Captura.read()
    Diferencial = cv2.absdiff(Frame1, Frame2)
    ImagenDifGray = cv2.cvtColor(Diferencial, cv2.COLOR_BGR2GRAY)
    DifuminadoDifGray = cv2.GaussianBlur(ImagenDifGray, (5,5), 0)
    _, BinarizadoDif = cv2.threshold(DifuminadoDifGray, 20, 255, cv2.THRESH_BINARY)
    DilatadoDif = cv2.dilate(BinarizadoDif, None, iterations=3)
    Contornos, _ = cv2.findContours(DilatadoDif, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    for Contorno in Contornos:
        (x, y, w, h) = cv2.boundingRect(Contorno)

        if cv2.contourArea(Contorno) < 900:
            continue
        
        cv2.rectangle(Frame1, (x,y), (x+w, y+h), (0,255,0), 2)
    
    Imagen = cv2.resize(Frame1, (1280,720))
    VideoSalida.write(Imagen)
    Frame1 = Frame2
    Estado, Frame2 = Captura.read()

    if cv2.waitKey(40) == 27:
        break

Captura.release()
VideoSalida.release()


    
````
